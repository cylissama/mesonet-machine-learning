```python
# im creating an 'inverse' type file here that includes a binary representation of where the exists data (0) vs no data (1)
# this should make it easier to query  for where missing data is.
# this is indexed via the original files timestamp values
import pandas as pd
import numpy as np

CRRL_file = '../CRRL.csv'
RFSM_file = '../RFSM.csv'

CRRL_df = pd.read_csv(CRRL_file)

RFSM_df = pd.read_csv(RFSM_file)
# Read the original CSV file
CRRL_df_original = CRRL_df.copy()

RFSM_df_original = RFSM_df.copy()

CRRL_df_original_indexed = CRRL_df_original.set_index('UTCTimestampCollected')

RFSM_df_original_indexed = RFSM_df_original.set_index('UTCTimestampCollected')
# Create the inverse DataFrame
# This converts NaN to 1 and non-NaN to 0
CRRL_df_inverse = CRRL_df_original_indexed.isna().astype(int)
RFSM_df_inverse = RFSM_df_original_indexed.isna().astype(int)

# Display the results
print("Original DataFrame shape:", CRRL_df_original.shape)
print("Inverse DataFrame shape:", CRRL_df_inverse.shape)

print("\nOriginal DataFrame (first 5 rows):")
print(CRRL_df_original.head())

print("\nInverse DataFrame (first 5 rows):")
print(CRRL_df_inverse.head())

# Save the inverse DataFrame to a new CSV file
CRRL_df_inverse.to_csv('missing_values_map.csv', index=False)

print("\nInverse DataFrame saved to 'missing_values_map.csv'")

# Optional: Show summary statistics
print(f"\nSummary:")
print(f"Total cells in original: {CRRL_df_original.size}")
print(f"Total missing values: {CRRL_df_inverse.sum().sum()}")
print(f"Percentage missing: {(CRRL_df_inverse.sum().sum() / CRRL_df_original.size) * 100:.2f}%")
```
