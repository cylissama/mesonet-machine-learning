```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files into DataFrames
file1 = '/Users/cylis/Work/mes_summer25/CRRL mods/CRRL_mesonet_data_filled_v2.csv'
file2 = '/Users/cylis/Work/mes_summer25/original/CRRL.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Ensure both DataFrames have the same columns
common_columns = df1.columns.intersection(df2.columns)

# Loop through each common column and plot
for column in common_columns:
    if pd.api.types.is_numeric_dtype(df1[column]) and pd.api.types.is_numeric_dtype(df2[column]):
        plt.figure(figsize=(10, 5))
        plt.plot(df1[column], label=f'{file1} - {column}', color='blue')
        plt.plot(df2[column], label=f'{file2} - {column}', color='orange')
        plt.title(f'Comparison of "{column}" between the two files')
        plt.xlabel('Index')
        plt.ylabel(column)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```
