```python
# this creates a separate file for each site without the soil data

import pandas as pd
import numpy as np

CRRL_file = '../CRRL.csv'
RFSM_file = '../RFSM.csv'

CRRL_df = pd.read_csv(CRRL_file)
RFSM_df = pd.read_csv(RFSM_file)

CRRL_no_soil = CRRL_df.drop(columns=['SM02','SM04','ST02','ST04'])
RFSM_no_soil = RFSM_df.drop(columns=['SM02','SM04','ST02','ST04'])

target_vars = ['TAIR','VT90','SM02','PRES']
target_vars_no_soil = ['TAIR','VT90','PRES'] # since no soil has SM02 dropped I removed it for simplicity

CRRL_subset = CRRL_df.dropna(subset=target_vars)
RFSM_subset = RFSM_df.dropna(subset=target_vars) # this has nothing cause of missing data
CRRL_subset_no_soil = CRRL_no_soil.dropna(subset=target_vars_no_soil)
RFSM_subset_no_soil = RFSM_no_soil.dropna(subset=target_vars_no_soil)

# print(CRRL_subset.shape)
# print(RFSM_subset.shape)
# print(CRRL_subset_no_soil.shape)
# print(RFSM_subset_no_soil.shape)
```

```python
print("\nCRRL w/ soil",
      "\nStart:", CRRL_subset['UTCTimestampCollected'].iloc[1],
      "\nEnd:", CRRL_subset['UTCTimestampCollected'].iloc[-1])
print("\nRFSM w/ soil",
      "\nStart:", RFSM_subset['UTCTimestampCollected'].iloc[0], # no data here
      "\nEnd:", RFSM_subset['UTCTimestampCollected'].iloc[-1])

print("\nCRRL no soil",
      "\nStart:", CRRL_subset_no_soil['UTCTimestampCollected'].iloc[1],
      "\nEnd:", CRRL_subset_no_soil['UTCTimestampCollected'].iloc[-1])
print("\nRFSM no soil",
      "\nStart:", RFSM_subset_no_soil['UTCTimestampCollected'].iloc[1], # no data here
      "\nEnd:", RFSM_subset_no_soil['UTCTimestampCollected'].iloc[-1])
```
